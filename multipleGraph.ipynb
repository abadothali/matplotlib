Multiple Graph...

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

import data from excel sheet...

```python
p=pd.read_excel('data.xlsx')
print(p)
```

Multiple graph....

```python
plt.plot(p.Years,p.India,'b*--',label='India')
plt.plot(p.Years,p.Bangladesh,'r*-.',label='Bangladesh')
plt.plot(p.Years,p.Dubai,'g*-',label="Dubai")

plt.xlabel('Years')
plt.ylabel('Price')
plt.title('Mobile price',fontdict={'fontname':'Comic Sans MS','fontsize':14})
plt.legend()
plt.savefig('price.png',dpi=200)
plt.show()
```
