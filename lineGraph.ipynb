Import matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X=[5,10,15,20,25]
Y=[4,8,12,16,20]
p=plt.plot(X,Y)
print(p)
```

Graph color change

```python
X=[5,10,15,20,25]
Y=[4,8,12,16,20]
p=plt.plot(X,Y,color='red')
print(p)
```

If only show

```python
X=[5,10,15,20,25]
Y=[4,8,12,16,20]
plt.plot(X,Y)
plt.show()
```

For marker...

```python
X=[5,10,15,20,25]
Y=[4,8,12,16,20]
plt.plot(X,Y,marker='*')
plt.show()
```

We can change also marker color and size...

```python
X=[5,10,15,20,25]
Y=[4,8,12,16,20]
p=plt.plot(X,Y,color='red',marker='*',markeredgecolor='blue',markersize=15)
print(p)
```

We can change also line style....

```python
X=[5,10,15,20,25]
Y=[4,8,12,16,20]
p=plt.plot(X,Y,color='red',marker='*',markeredgecolor='blue',linestyle='--')
print(p)
```

we can create this is a short forms...

```python
X=[5,10,15,20,25]
Y=[4,8,12,16,20]

# first line color, next marker, than line style (color, marker, line style)
p=plt.plot(X,Y,'b*--')
print(p)
```

For heading the graph...

```python
X=[5,10,15,20,25]
Y=[4,8,12,16,20]
plt.plot(X,Y,color='red',marker='*',markeredgecolor='blue',linestyle='--')
plt.title('Basic Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})
```

For X-Axis Or Y-Axis

```python
X=[5,10,15,20,25]
Y=[4,8,12,16,20]
plt.plot(X,Y,'b*--')

# For title...
plt.title('Basic Graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})

# for X asix & Y asix
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')

# for additional informations...
plt.legend(['Single Elements..'])

plt.show()
```

For save figer....

```python
X=[5,10,15,20,25]
Y=[4,8,12,16,20]
plt.plot(X,Y)
plt.savefig('firstfig.png',dpi=200)
plt.show()
```
