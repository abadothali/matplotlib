```python
import matplotlib.pyplot as plt
```

single scatter plot

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.scatter(x,y)
plt.show()
```

multiple scatter plot

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]

a=[11,12,13,14,15]
b=[6,7,8,9,10]

p=[9,8,13,22,21]
q=[16,17,8,19,10]

plt.scatter(x,y)
plt.scatter(p,q)
plt.scatter(a,b)

plt.show()
```
