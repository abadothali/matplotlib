```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Basic boxplot

```python
test=np.random.normal(200,50,400)
plt.boxplot(test)
plt.show()
```

Multiple boxplot

```python
test1=np.random.normal(200,50,400)
test2=np.random.normal(100,150,350)
test3=np.random.normal(200,70,270)
test4=np.random.normal(250,100,400)

data=[test1,test2,test3,test4]
plt.boxplot(data)
plt.show()
```
