Piechart....

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# pie chart...
data=np.array([15,30,35,40])
name=['Ajay','vijay','ram','shyam']

plt.pie(data,labels=name)
plt.show()
```

```python
# We can also....
data=[15,30,35,40]
name=['Ajay','vijay','ram','shyam']

plt.pie(data,labels=name)
plt.show()
```

Explode....

```python
name_explode=[0,0,0.2,0]
plt.pie(data, labels=name,explode=name_explode)
plt.show()
```

Shadow...

```python
name_explode=[0,0,0.2,0]
plt.pie(data, labels=name,explode=name_explode, shadow=True)
plt.show()
```

Save figer...

```python
name_explode=[0,0,0.2,0]
plt.pie(data, labels=name,explode=name_explode, shadow=True)
plt.savefig('pie.png',dpi=200)
plt.show()
```

Color....

```python
# color...
mycolor=['b','y','g','m']

# explode...
nameexplode=[0,0,0,0.1]

# shadow....
plt.pie(data,labels=name,colors=mycolor,explode=nameexplode,shadow=True)

# legend...
plt.legend()

plt.show()
```
